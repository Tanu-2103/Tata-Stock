## Hypothesis

## z-testing

```python
# Calculate the average open price of Tata Steel stock
open_price = np.mean(tatastock['Open Price'])

# Calculate the average close price of Tata Steel stock
close_price = np.mean(tatastock['Close Price'])

# Calculate the standard deviation of Tata Steel open prices
open_price_std = np.std(tatastock['Open Price'])

# Calculate the standard deviation of Tata Steel close prices
close_price_std = np.std(tatastock['Close Price'])

# Calculate the t-statistic
t_statistic = (open_price - close_price) / np.sqrt(open_price_std**2 + close_price_std**2)

# Calculate the p-value
p_value = stats.ttest_ind(tatastock['Open Price'], tatastock['Close Price'], equal_var=False).pvalue

# If the p-value is less than the significance level, then reject the null hypothesis
if p_value < 0.05:
    print('Reject the null hypothesis')
else:
    print('Accept the null hypothesis')
```

## chi-testing

```python
# Create a contingency table
contingency_table = tatastock[['No.of Shares', 'Total Turnover (Rs.)']]
contingency_table = contingency_table.groupby(['No.of Shares', 'Total Turnover (Rs.)']).size().unstack()

# Calculate the chi-squared statistic and p-value
chi2_statistic, p_value, dof, expected_table = chi2_contingency(contingency_table)

# If the p-value is less than the significance level, then reject the null hypothesis
if p_value < 0.05:
    print('Reject the null hypothesis')
else:
    print('Accept the null hypothesis')
```

## Outliers

```python
def find_outliers_iqr(data):
    # Calculate the quartiles
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    # Calculate the interquartile range
    IQR = Q3 - Q1
    # Calculate the upper and lower fences
    upper_fence = Q3 + 1.5 * IQR
    lower_fence = Q1 - 1.5 * IQR
    outlier= np.where((data > upper_fence) | (data < lower_fence))[0]
    return outlier



# Find the outliers in the opening price data
outliers = find_outliers_iqr(tatastock['Close Price'])

# Print the indices of the outliers
print(outliers)
```

```python
# Create a boxplot of the opening price data
sb.boxplot(
    x='Close Price',
    showmeans=True,
    data=tatastock,
)

# Mark the outliers in red
sb.swarmplot(
    x='Close Price',
    data=tatastock.iloc[outliers],
    color='red',
)

# Show the plot
plt.show()
```

### Is the data balanced or imbalanced?

```python
import numpy as np

def calculate_class_imbalance_ratio(labels):

  # Get the unique labels
    unique_labels = np.unique(labels)

  # Calculate the number of samples in each class
    num_samples_per_class = np.array([np.sum(labels == label) for label in unique_labels])

  # Calculate the class imbalance ratio
    class_imbalance_ratio = np.max(num_samples_per_class) / np.min(num_samples_per_class)
    return class_imbalance_ratio



# Calculate the class imbalance ratio of the dataset
class_imbalance_ratio = calculate_class_imbalance_ratio(tatastock['Total Turnover (Rs.)'])

# Print the class imbalance ratio
# Calculate the proportion of each class in the dataset
class_proportions = tatastock['Total Turnover (Rs.)'].value_counts(normalize=True)

# Create a histogram of the class imbalance ratio
plt.hist(class_imbalance_ratio)
plt.xlabel("Class Imbalance Ratio")
plt.ylabel("Frequency")
plt.title("Histogram of Class Imbalance Ratio in Tata Steel Dataset")
plt.show()
```
